```python
#Bài 1: Nhập một số tự nhiên N. Vẽ hình dấu sao theo quy luật như ví dụ bên dưới với N = 5:

n= 7

for i in range(2*n-1):
    if i%2==0:
        if i <= (n-1):
            a = i+1
            b = int((n-a)/2)
        else:
            a = 2*n-1-i
            b = int((n-a)/2)
        print(str(' '*b) + str('*'*a) + str(' '*b))
```

```python
#Bài 2: Cho một tờ tiền mệnh giá 100 ngàn. 
#Hỏi có tối đa bao nhiêu cách đổi tờ tiền đó ra các tờ có mệnh giá nhỏ hơn là 2 ngàn, 3 ngàn và 5 ngàn.

T = 100 #100 nghìn đồng
a = 2 # 2 nghìn đồng
b = 3 # 3 nghìn đồng
c = 5 # 5 nghìn đồng
x = T//a + 1 
y = T//b + 1 
z = T//c + 1 
dem = 0
for i in range(x):
    for j in range(y):
        for k in range(z):
            if i*a + j*b + k*c==T:
                dem += 1
                print('Cách {0}: \n {1}*{2} + {3}*{4} + {5}*{6} ='.format(dem, i, a, j, b, k, c), i*a+j*b+k*c)
print()
print('=> Có {0} cách đổi tờ 100 nghìn sang tờ 2 nghìn, 3 nghìn, 5 nghìn'.format(dem))
            
            
```
